```python
#How many are there in each Team and the percentage splitting with respect to the total employees.
df['Team'].value_counts()
```

```python
#Percetage of employees in each group with respect to total employees
percentage_split = (df['Team'].value_counts()/len(df))*100
percentage = percentage_split.round(2)
percentageDF = pd.DataFrame({'Team':df['Team'].value_counts(),'percentage split':percentage})
percentageDF
```

```python
percentage_split = (df['Team'].value_counts()/len(df))*100
percentage = percentage_split.round(2)
percentageDF = pd.DataFrame({'Team':df['Team'].value_counts(),'percentage split':percentage})
percentageDF
```

```python

plt.figure(figsize=(15, 8))
sns.barplot(x='Team', y='percentage split',data=percentageDF,palette='tab10',hue=percentageDF['Team'].index)
plt.title('Percentage split')
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

# File path to the CSV file
file_path = r"C:\Users\abhin\Downloads\abc-company.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Ensure the column name 'Position' matches the one in your CSV file
if 'Position' not in df.columns:
    raise ValueError("'Position' column is not in the DataFrame")

# Count the occurrences of each position
position_counts = df['Position'].value_counts()

# Print the results
print(position_counts)
```

```python
df[['Position','Name']]
```

```python

groupby_position = df.groupby('Position')['Name']
```

```python
for position, name in groupby_position:
    print(f"Position: {position}")
    print(f"name: {name}")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Position',data=df,palette='rocket',hue='Position')
plt.title("employee position")
```

```python

df[['Team','Position','Salary']]
```

```python
total_salary_spend = df.groupby(['Team','Position'])['Salary'].sum()
max_salary = total_salary_spend.max()
max_salary_index = total_salary_spend.idxmax()
max_salary_index
max_salary
print(f"maximum salary spend {max_salary}, spend by {max_salary_index}")
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df,x=df['Team'],y=df['Salary'],hue=df['Position'])
plt.xticks(rotation=90)
plt.show()
```
